```python
import requests
from bs4 import BeautifulSoup

# URL страницы
url = "https://www.cbr.ru/press/pr/?file=638600287613725155BANK_SECTOR.htm"

# Отправка запроса на страницу
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Поиск нужного div (например, по классу)
div_container = soup.find('div', class_='note')

# Если такой div найден, извлекаем все абзацы внутри него
if div_container:
    paragraphs = div_container.find_all('p')
    for paragraph in paragraphs:
        print(paragraph.text)
```

Пример №2

```python
import requests
from bs4 import BeautifulSoup

# URL страницы
url = "https://ifcongress.ru/2024videos/ru"

# Отправка запроса на страницу
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Поиск всех ссылок на видео
video_links = []

# Ссылки на видео находятся внутри тегов <a> с атрибутом onclick
for a_tag in soup.find_all('a', onclick=True):
    onclick_value = a_tag['onclick']
    if "showVideoFile2024" in onclick_value:
        # Извлечение URL из строки
        start = onclick_value.find('(') + 2
        end = onclick_value.find(',', start) - 1
        video_url = onclick_value[start:end]
        video_links.append(video_url)

# Вывод всех найденных ссылок
for link in video_links:
    print(link)
```

beautifulsoup

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

# Базовый URL для получения курсов валют
base_url = 'https://www.cbr.ru/currency_base/daily/'

# Получение текущей даты
end_date = datetime.now()

# Вычисление даты месяц назад
start_date = end_date - timedelta(days=30)

# Формат даты, используемый на сайте ЦБ РФ
date_format = '%d.%m.%Y'

# Цикл по всем дням в пределах последнего месяца
current_date = start_date
while current_date <= end_date:
    # Формируем URL с нужной датой
    date_str = current_date.strftime(date_format)
    url = f"{base_url}?UniDbQuery.Posted=True&UniDbQuery.To={date_str}"

    # Отправка GET-запроса на сайт
    response = requests.get(url)

    # Проверка статуса ответа
    if response.status_code == 200:
        # Парсинг HTML-кода с использованием BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Поиск таблицы с курсами валют
        currency_table = soup.find('table', {'class': 'data'})

        if currency_table:
            print(f"Курсы валют на {date_str}:")
            # Проход по всем строкам таблицы
            rows = currency_table.find_all('tr')[1:]  # Пропускаем заголовок таблицы
            for row in rows:
                cols = row.find_all('td')
                currency_name = cols[1].text.strip()
                currency_rate = cols[4].text.strip()
                print(f"  Валюта: {currency_name}, Курс: {currency_rate}")
            print("-" * 40)
        else:
            print(f"Таблица с курсами валют не найдена на {date_str}.")
    else:
        print(f"Не удалось получить доступ к странице на {date_str}. Код ответа: {response.status_code}")

    # Переход к следующему дню
    current_date += timedelta(days=1)
```

lxml

```python
import requests
from lxml import html
import pandas as pd
from datetime import datetime, timedelta

# Функция для формирования URL на основе выбранной даты
def generate_url(start_date, end_date):
    base_url = 'https://www.cbr.ru/hd_base/ostat_depo_new/'
    return f'{base_url}?UniDbQuery.Posted=True&UniDbQuery.From={start_date}&UniDbQuery.To={end_date}'

# Ввод начальной и конечной даты пользователем
start_date = input("Введите начальную дату (в формате ДД.ММ.ГГГГ): ")
end_date = input("Введите конечную дату (в формате ДД.ММ.ГГГГ): ")

# Формируем URL на основе введенных дат
url = generate_url(start_date, end_date)

# Отправка GET-запроса на сайт
response = requests.get(url)

# Проверка статуса ответа
if response.status_code == 200:
    # Парсинг HTML-кода с использованием lxml
    tree = html.fromstring(response.content)

    # Заголовки таблицы (извлекаем текст из <th>)
    headers = tree.xpath('//table[@class="data spaced"]//th/text()')
    headers = [header.strip() for header in headers if header.strip()]
    print("Заголовки:", headers)

    # Извлечение строк таблицы
    rows = tree.xpath('//table[@class="data spaced"]//tr[td]')

    # Сохранение данных в список словарей
    data = []
    for row in rows:
        # Извлечение данных из ячеек (td)
        cells = row.xpath('.//td/text()')
        cells = [cell.strip() for cell in cells if cell.strip()]
        # Создаем словарь только если количество ячеек соответствует количеству заголовков
        if len(cells) == len(headers):
            data.append({headers[i]: cells[i] for i in range(len(cells))})
        else:
            # Дополнительная обработка случаев, когда ячеек меньше, чем заголовков
            temp_dict = {}
            for i in range(len(cells)):
                temp_dict[headers[i]] = cells[i]
            data.append(temp_dict)

    # Преобразование данных в DataFrame для удобства анализа
    df = pd.DataFrame(data)
    print(df)

    # Сохранение в CSV файл
    df.to_csv(f'deposit_balances_{start_date}_to_{end_date}.csv', index=False, encoding='utf-8-sig')

else:
    print(f"Не удалось получить доступ к странице. Код ответа: {response.status_code}")
```

# Парсинг Телеграма



```python
import asyncio
import nest_asyncio
import os
from telethon import TelegramClient
from telethon.tl.functions.messages import GetHistoryRequest
from getpass import getpass

# Применение nest_asyncio для поддержки вложенных циклов событий
nest_asyncio.apply()

# Ввод данных с использованием getpass
api_id = int(getpass("Введите ваш api_id: "))
api_hash = getpass("Введите ваш api_hash: ")
phone = getpass("Введите ваш номер телефона: ")

# Удаление старого файла сессии, если он существует
session_file = 'session_name.session'
if os.path.exists(session_file):
    os.remove(session_file)

async def main():
    client = TelegramClient(session_file, api_id, api_hash)
    await client.start()

    # Запрашиваем у пользователя название или ссылку на Telegram-канал
    channel_name = input("Введите название или ссылку на Telegram-канал (например, @example_channel): ")

    # Получаем сущность канала
    target_channel = await client.get_entity(channel_name)

    # Парсинг последних 10 сообщений
    offset_id = 0
    limit = 10
    all_messages = []

    history = await client(GetHistoryRequest(
        peer=target_channel,
        offset_id=offset_id,
        offset_date=None,
        add_offset=0,
        limit=limit,
        max_id=0,
        min_id=0,
        hash=0
    ))

    messages = history.messages
    for message in messages:
        # Сохраняем только текст сообщений
        if message.message:  # проверка, что сообщение не пустое
            all_messages.append(message.message.strip())

    print("Сохраняем сообщения в файл...")

    # Сохраняем текст сообщений в текстовый файл
    with open("channel_messages.txt", "w", encoding="utf-8") as f:
        for message in all_messages:
            f.write(message + "\n")

    print('Парсинг последних 10 сообщений канала успешно выполнен.')

# Запуск асинхронной функции в Jupyter Notebook
await main()
```
